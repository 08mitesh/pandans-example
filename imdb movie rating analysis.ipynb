```python
import pandas as pd
df = pd.read_csv('movie_metadata.csv')
df.head().T
```

```python
#simple crosstab with single column
pd.crosstab(df['director_name'], df['country'])
```

```python
#Pandas crosstabe can be used also for multiple columns for rows or columns. The syntax for multiple rows can be seen below:
pd.crosstab([df['director_name'],df['genres']], df['country'])
```

```python
#Use values from another column and aggregation function
import numpy as np
pd.crosstab(df['director_name'], df['country'], values=df.imdb_score, aggfunc=np.sum)
```
